# Reading Training Data

```python
training_word_set = set(open('training.txt').read().split())
print(training_word_set)
```

# Importing Trie Data Structure (Prefix Tree)

https://pythonhosted.org/PyTrie/

```python
# conda activate python3
# pip install pytrie
from pytrie import SortedStringTrie as trie
```

# Transform to Lower Case

```python
training_word_set = map(lambda x: x.lower(), training_word_set)
```

# Populating Prefix Tree

```python
prefix_tree = trie.fromkeys(training_word_set)
print(prefix_tree)
```

# Prefix Tree Keys

```python
prefix_tree.keys()
```

# Intellisense Functions 

```python
prefix_tree.keys(prefix = str.lower('The'))
```

```python
print(prefix_tree.longest_prefix('african',default = -1))
```

```python
list(prefix_tree.iter_prefixes('programming'))
```
